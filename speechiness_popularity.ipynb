```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "spotify_songs.csv"
df = pd.read_csv(file_path)
print(df.head())
```

```python
print(df.shape)
df = df.dropna()
df = df.drop_duplicates(subset="track_id")
print(df.shape)
```

```python
relevant_columns = [
    "track_popularity",
    "danceability",
    "energy",
    "key",
    "loudness",
    "mode",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
]

df = df[relevant_columns]
```

```python
df_dance = df.groupby("danceability")["track_popularity"].mean().reset_index()
df_energy = df.groupby("energy")["track_popularity"].mean().reset_index()
df_key = df.groupby("key")["track_popularity"].mean().reset_index()
df_loud = df.groupby("loudness")["track_popularity"].mean().reset_index()
df_mode = df.groupby("mode")["track_popularity"].mean().reset_index()
df_speech = df.groupby("speechiness")["track_popularity"].mean().reset_index()
df_acoust = df.groupby("acousticness")["track_popularity"].mean().reset_index()
df_instru = (
    df.groupby("instrumentalness")["track_popularity"].mean().reset_index()
)
df_live = df.groupby("liveness")["track_popularity"].mean().reset_index()
df_valence = df.groupby("valence")["track_popularity"].mean().reset_index()
```
